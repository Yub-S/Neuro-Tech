```python
import streamlit as st
import mysql.connector
from datetime import datetime
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from PIL import Image
import os

# Function to process the image and get report data
def process_image(image_path):
    # Load the Google GenAI model
    API_KEY = "AIzaSyDqOnKHs4JoKXHjgLU4JeNqcbwplDioRSs"
    os.environ["GOOGLE_API_KEY"] = API_KEY
    llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash", api_key=os.environ["GOOGLE_API_KEY"])
    
    message = HumanMessage(
        content=[
            {"type": "text", "text": "return every report parameters in dictionary format"},
            {"type": "image_url", "image_url": {"url": image_path}},
        ],
    )
    
    response = llm.invoke([message])
    result = dict(response.content)
    return result

# Function to upload data to MySQL
def upload_to_mysql(data):
    # Convert string dates to datetime objects
    collection_date = datetime.strptime(data['CollectionDate'], '%d/%m/%Y %I:%M %p')
    report_date = datetime.strptime(data['ReportDate'], '%d/%m/%Y %I:%M %p')

    # Connect to your local MySQL server
    conn = mysql.connector.connect(
        host='localhost',
        user='root',
        password='your_mysql_password',
        database='MedicalReports'
    )
    cursor = conn.cursor()

    # Insert data into the table
    query = """
    INSERT INTO PatientReport (
        Name, AgeGender, ReferredBy, PhoneNo, PatientID, ReportID,
        CollectionDate, ReportDate, Haemoglobin, TotalLeucocyteCount,
        DifferentialLeucocyteCount, Neutrophils, Lymphocytes, Eosinophils,
        Monocytes, Basophils, AbsoluteLeucocyteCount, AbsoluteNeutrophils,
        AbsoluteLymphocytes, AbsoluteEosinophils, AbsoluteMonocytes, RBCCount,
        MCV, MCH, MCHC, Hct, RDWCV, RDWSD, PlateletCount, PCT, MPV, PDW,
        Interpretation
    ) VALUES (
        %(Name)s, %(AgeGender)s, %(ReferredBy)s, %(PhoneNo)s, %(PatientID)s,
        %(ReportID)s, %(CollectionDate)s, %(ReportDate)s, %(Haemoglobin)s,
        %(TotalLeucocyteCount)s, %(DifferentialLeucocyteCount)s, %(Neutrophils)s,
        %(Lymphocytes)s, %(Eosinophils)s, %(Monocytes)s, %(Basophils)s,
        %(AbsoluteLeucocyteCount)s, %(AbsoluteNeutrophils)s, %(AbsoluteLymphocytes)s,
        %(AbsoluteEosinophils)s, %(AbsoluteMonocytes)s, %(RBCCount)s, %(MCV)s,
        %(MCH)s, %(MCHC)s, %(Hct)s, %(RDWCV)s, %(RDWSD)s, %(PlateletCount)s,
        %(PCT)s, %(MPV)s, %(PDW)s, %(Interpretation)s
    )
    """

    cursor.execute(query, {
        'Name': data['Name'],
        'AgeGender': data['AgeGender'],
        'ReferredBy': data['ReferredBy'],
        'PhoneNo': data['PhoneNo'],
        'PatientID': data['PatientID'],
        'ReportID': data['ReportID'],
        'CollectionDate': collection_date,
        'ReportDate': report_date,
        'Haemoglobin': data['Haemoglobin'],
        'TotalLeucocyteCount': data['TotalLeucocyteCount'],
        'DifferentialLeucocyteCount': data['DifferentialLeucocyteCount'],
        'Neutrophils': data['Neutrophils'],
        'Lymphocytes': data['Lymphocytes'],
        'Eosinophils': data['Eosinophils'],
        'Monocytes': data['Monocytes'],
        'Basophils': data['Basophils'],
        'AbsoluteLeucocyteCount': data['AbsoluteLeucocyteCount'],
        'AbsoluteNeutrophils': data['AbsoluteNeutrophils'],
        'AbsoluteLymphocytes': data['AbsoluteLymphocytes'],
        'AbsoluteEosinophils': data['AbsoluteEosinophils'],
        'AbsoluteMonocytes': data['AbsoluteMonocytes'],
        'RBCCount': data['RBCCount'],
        'MCV': data['MCV'],
        'MCH': data['MCH'],
        'MCHC': data['MCHC'],
        'Hct': data['Hct'],
        'RDWCV': data['RDWCV'],
        'RDWSD': data['RDWSD'],
        'PlateletCount': data['PlateletCount'],
        'PCT': data['PCT'],
        'MPV': data['MPV'],
        'PDW': data['PDW'],
        'Interpretation': data['Interpretation']
    })

    conn.commit()
    cursor.close()
    conn.close()

# Streamlit app
st.title("Medical Report Upload")

# File uploader
uploaded_file = st.file_uploader("Choose an image file", type=["jpg", "png", "jpeg"])

if uploaded_file is not None:
    # Display the image
    image = Image.open(uploaded_file)
    st.image(image, caption='Uploaded Image.', use_column_width=True)

    # Save the uploaded image
    image_path = os.path.join("temp", uploaded_file.name)
    image.save(image_path)

    # Process the image to get report data
    with st.spinner('Processing...'):
        report_data = process_image(image_path)
    
    st.success('Processing done!')
    st.write("Extracted Report Data:", report_data)

    # Upload the data to MySQL
    if st.button('Upload to Database'):
        with st.spinner('Uploading to database...'):
            upload_to_mysql(report_data)
        st.success('Data uploaded to database!')

if __name__ == '__main__':
    # Run the Streamlit app
    st.run()

```
